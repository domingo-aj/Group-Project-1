```python
import pandas as pd
```

```python
# Because of the size of the datasets and difficulty sharing them on GitHub,
# the team decided to limit data to 5 select airports from 2015-2018
year = 2015
focused_airports = ['ATL', 'LAX', 'ORD', 'DFW', 'DEN']
file = "raw_data/" + str(year) + ".csv"
```

```python
original_df = pd.read_csv(file, sep=",", header=0, index_col=False)

original_df
```

```python
# Search for null values that don't make sense and see column names.
original_df.isna().sum()
```

```python
filtered_df = original_df.iloc[:,[0,3,4,5,6,7,12,13,14,16,21,22,23,24,25,26]]
filtered_df = filtered_df.loc[original_df['ORIGIN'].isin(focused_airports), :]
filtered_df
```

```python
# Simplify cancellations into weather and non-weather
original_df['CANCELLATION_CODE'].unique()
```

```python
filtered_df['Non Weather Delay'] = filtered_df['CARRIER_DELAY']  + filtered_df['NAS_DELAY'] + \
                                   filtered_df['SECURITY_DELAY'] + filtered_df['LATE_AIRCRAFT_DELAY']

filtered_df = filtered_df.drop(columns=['CARRIER_DELAY', 'NAS_DELAY', 'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY'])

filtered_df.loc[filtered_df['CANCELLATION_CODE'] == 'B', 'WEATHER_DELAY'] = 'CANCELLED'
filtered_df.loc[(filtered_df['CANCELLATION_CODE'] == 'A') |
                (filtered_df['CANCELLATION_CODE'] == 'C') |
                (filtered_df['CANCELLATION_CODE'] == 'D'), 'Non Weather Delay'] = 'CANCELLED'

filtered_df = filtered_df.drop(columns = 'CANCELLATION_CODE')
```

```python
filtered_df
```

```python
# Give columns more intuitive names
filtered_df = filtered_df.rename(columns={'FL_DATE' : 'Date',
                                    'ORIGIN'        : 'Origin',
                                    'DEST'          : 'Destination',
                                    'CRS_DEP_TIME'  : 'Expected Departure Time',
                                    'DEP_TIME'      : 'Actual Departure Time',
                                    'DEP_DELAY'     : 'Departure Delay',
                                    'CRS_ARR_TIME'  : 'Expected Arrival Time',
                                    'ARR_TIME'      : 'Arrival Time',
                                    'ARR_DELAY'     : 'Arrival Delay',
                                    'DISTANCE'      : 'Distance',
                                    'WEATHER_DELAY' : 'Weather Delay'})
```

```python
# Interpret NaNs in delay rows as 0s.
filtered_df.iloc[:, [5,8,9,10,11]] = filtered_df.iloc[:, [5,8,9,10,11]].fillna(0)
```

```python
filtered_df
```

```python
# Data files were too big when combined so leave them year-by-year
output_path = 'clean_data/' + str(year) + 'test.csv'
filtered_df.to_csv(output_path, index=False)
```
