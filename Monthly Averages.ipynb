```python
import pandas as pd
from pprint import pprint
import time
```

```python
daily_df = pd.read_csv('airport_weather_data_result.csv')
daily_df = daily_df.drop(columns="Unnamed: 0")
daily_df
```

```python
daily_date = daily_df.set_index('Date', inplace=True)
daily_date = daily_df.index = pd.to_datetime(daily_df.index)

daily_df
```

```python
dfw_dailies = daily_df.loc[daily_df['Airport'] == "DFW"]
atl_dailies = daily_df.loc[daily_df['Airport'] == "ATL"]
den_dailies = daily_df.loc[daily_df['Airport'] == "DEN"]
lax_dailies = daily_df.loc[daily_df['Airport'] == "LAX"]
ord_dailies = daily_df.loc[daily_df['Airport'] == "ORD"]
```

```python
dfw_monthly_avg = dfw_dailies.resample('1M').mean()
dfw_monthly_avg['Airport'] = "DFW"

dfw_monthly_avg = dfw_monthly_avg.reset_index(drop=False)

dfw_monthly_avg.head()
```

```python
atl_monthly_avg = atl_dailies.resample('1M').mean()
atl_monthly_avg['Airport'] = "ATL"

atl_monthly_avg = atl_monthly_avg.reset_index(drop=False)

atl_monthly_avg.head()
```

```python
den_monthly_avg = den_dailies.resample('1M').mean()
den_monthly_avg['Airport'] = "DEN"

den_monthly_avg = den_monthly_avg.reset_index(drop=False)

den_monthly_avg.head()
```

```python
lax_monthly_avg = lax_dailies.resample('1M').mean()
lax_monthly_avg['Airport'] = "LAX"

lax_monthly_avg = lax_monthly_avg.reset_index(drop=False)

lax_monthly_avg.head()
```

```python
ord_monthly_avg = ord_dailies.resample('1M').mean()
ord_monthly_avg['Airport'] = "ORD"

ord_monthly_avg = ord_monthly_avg.reset_index(drop=False)

ord_monthly_avg.head()
```

```python
months = pd.concat([dfw_monthly_avg, atl_monthly_avg, den_monthly_avg, lax_monthly_avg, ord_monthly_avg], axis=0)
months
```

```python
months.to_csv('Monthly_Weather_Averages.csv')
```
