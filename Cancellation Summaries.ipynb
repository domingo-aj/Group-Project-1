```python
import pandas as pd 
```

```python
summary2015 = pd.read_csv('Data_Cleaning/clean_data/focused_airports_2015.csv')
summary2016 = pd.read_csv('Data_Cleaning/clean_data/focused_airports_2016.csv')
summary2017 = pd.read_csv('Data_Cleaning/clean_data/focused_airports_2017.csv')
summary2018 = pd.read_csv('Data_Cleaning/clean_data/focused_airports_2018.csv')

summary1 = summary2015.append(summary2016)
summary2 = summary1.append(summary2017)
summary = summary2.append(summary2018)

total_flights = len(summary)
total_cancels = len(summary.loc[summary["Weather Delay"] == "CANCELLED"])
```

```python
#total by month: cancels, non-cancels, total flights

#summary by airport: 

#Airport #total number of cancels #total number of non-cancelled #total flights
```

```python
summary_date = summary.set_index('Date', inplace=True)
summary_date = summary.index = pd.to_datetime(summary.index)
summary
```

```python
non_cancelled = summary.loc[summary["Weather Delay"] != "CANCELLED"]
cancelled = summary.loc[summary["Weather Delay"] == "CANCELLED"]
```

```python
cancelled
```

```python
resample_c = cancelled.resample('1M').count()
resample_nc = non_cancelled.resample('1M').count()
```

```python
monthly_cancels = resample_c["Weather Delay"].reset_index()
monthly_nc = resample_nc["Weather Delay"].reset_index()
```

```python
monthly_summary = pd.merge(monthly_cancels, monthly_nc, on="Date")
monthly_summary = monthly_summary.rename(columns={"Weather Delay_x": "# Cancelled Flights", 
                                                 "Weather Delay_y": "# Non-Cancelled Flights"})
monthly_summary["Total Flights"] = monthly_summary["# Cancelled Flights"] + monthly_summary["# Non-Cancelled Flights"]
```

```python
monthly_summary
```

```python
dfw = summary.loc[summary["Origin"] == "DFW"]
atl = summary.loc[summary["Origin"] == "ATL"]
den = summary.loc[summary["Origin"] == "DEN"]
lax = summary.loc[summary["Origin"] == "LAX"]
oha = summary.loc[summary["Origin"] == "ORD"]
```

```python
dfw_total = len(dfw)
atl_total = len(atl)
den_total = len(den)
lax_total = len(lax)
ord_total = len(oha)

dfw_cancels = dfw.loc[dfw["Weather Delay"] == "CANCELLED"]
atl_cancels = atl.loc[atl["Weather Delay"] == "CANCELLED"]
den_cancels = den.loc[den["Weather Delay"] == "CANCELLED"]
lax_cancels = lax.loc[lax["Weather Delay"] == "CANCELLED"]
ord_cancels = oha.loc[oha["Weather Delay"] == "CANCELLED"]

dfw_nc = dfw.loc[dfw["Weather Delay"] != "CANCELLED"]
atl_nc = atl.loc[atl["Weather Delay"] != "CANCELLED"]
den_nc = den.loc[den["Weather Delay"] != "CANCELLED"]
lax_nc= lax.loc[lax["Weather Delay"] != "CANCELLED"]
ord_nc = oha.loc[oha["Weather Delay"] != "CANCELLED"]

dfw_nc_total = len(dfw_nc)
atl_nc_total = len(atl_nc)
den_nc_total = len(den_nc)
lax_nc_total = len(lax_nc)
ord_nc_total = len(ord_nc)

dfw_total_cancels = len(dfw_cancels)
atl_total_cancels = len(atl_cancels)
den_total_cancels = len(den_cancels)
lax_total_cancels = len(lax_cancels)
ord_total_cancels = len(ord_cancels)

airport_names = ["DFW", "ATL", "DEN", "LAX", "ORD"]
airport_totals = [dfw_total, atl_total, den_total, lax_total, ord_total]
airport_nc = [dfw_nc_total, atl_nc_total, den_nc_total, lax_nc_total, ord_nc_total]
airport_cancels = [dfw_total_cancels, atl_total_cancels, den_total_cancels, lax_total_cancels, ord_total_cancels]
```

```python
airport_summary = pd.DataFrame({"Airport": airport_names, 
                              "# Cancelled Flights": airport_cancels, 
                              "# Non-Cancelled Flights": airport_nc, 
                              "Total Flights": airport_totals})
```
